```python
import plotly.express as px
import json
from pathlib import Path
import numpy as np
from scipy.stats import sem
import pandas as pd
```

```python
num_iters = 3
PATH = '../saved_metrics/'

max_vars = []
random = []

for i in range(num_iters):
    with open(Path(PATH, f'random_iteration_{i}.json'), 'r') as f:
        random.append(json.load(f))

for i in range(num_iters):
    with open(Path(PATH, f'max_variance_iteration_{i}.json'), 'r') as f:
        max_vars.append(json.load(f))
```

```python
rand = np.array(random)

max_var = np.array(max_vars)
```

```python
data_dict = {'iterations': 2 * [i for i in range(len(random[0]))], 
             'mse': rand.mean(axis=0).tolist() + max_var.mean(axis=0).tolist(),
             'sd_error': (2 * sem(rand)).tolist() + (2 * sem(max_var)).tolist(),
             'type': ['random' for i in range(len(random[0]))] + ['maximum variance' for i in range(len(max_vars[0]))]}

df = pd.DataFrame(data_dict)

fig = px.line(df, x='iterations', y='mse', color='type', error_x='sd_error')
fig.show()
```
